### Summary Statistics

Quickly shows distribution for numeric columns (mean, std, min, max)

Shows unique counts, most frequent values for categorical columns.

Helps you detect anomalies (negative play counts or ages).


```python
df.describe(include='all')
```

```python
## missing value analysis
missing_values = df.isnull().sum().sort_values(ascending=False)
percentage_missing = df.isnull().sum() / len(df) * 100
print(percentage_missing)
```
